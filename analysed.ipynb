```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/home/bryan/my_notebooks/my-jupyter-env/lib/python3.10/site-packages/data.csv")

print("Shape of the dataset:", df.shape)
print("Missing values:", df.isnull().sum())
print("Statistics for the 'amount' column:\n", df["amount"].describe())

counts = df["type"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=counts.index, y=counts.values)
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.title("Count of Each Transaction Type")
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df["amount"], bins=50)
plt.xlabel("Transaction Amount")
plt.ylabel("Count")
plt.title("Distribution of Transaction Amounts")
plt.show()

fraud_counts = df["isFraud"].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(fraud_counts, labels=["Non-Fraudulent", "Fraudulent"], autopct="%1.1f%%")
plt.title("Proportion of Fraudulent vs. Non-Fraudulent Transactions")
plt.show()

fraud_by_type = df[df["isFraud"] == 1]["type"].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=fraud_by_type.index, y=fraud_by_type.values)
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.title("Number of Fraudulent Transactions by Type")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x="oldbalanceOrg", y="newbalanceOrig", data=df)
plt.xlabel("Old Balance (Origin)")
plt.ylabel("New Balance (Origin)")
plt.title("Scatter Plot of Old Balance vs. New Balance (Origin)")
plt.show()

flagged_fraud_counts = df["isFlaggedFraud"].value_counts()
plt.figure(figsize=(6, 6))
plt.bar(flagged_fraud_counts.index, flagged_fraud_counts.values)
plt.xlabel("Flagged Fraud")
plt.ylabel("Count")
plt.title("Number of Flagged Fraud Transactions")
plt.show()
```
