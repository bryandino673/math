```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,400)
```

```python
def g(x):
    return x**2 + 2*x +5
    
plt.plot(x,g(x))
plt.xlabel("x")
plt.ylabel("g(x)")
plt.title('basic quadratic function')
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-10,10,400)
```

```python
y=np.sin(x)
```

```python
def h(x):
    return y

plt.plot(x, h(x))
plt.title("sine function")
plt.xlabel("x")
plt.ylabel("h(x)")
plt.grid(True)
plt.show()
```

```python
x=np.linspace(-10,10,400)
```
